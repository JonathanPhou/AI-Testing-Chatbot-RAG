```python
from huggingface_hub import InferenceClient
import os
```

```python
api_token = os.getenv("HF_API_KEY")
model_name = "openai/gpt-oss-120b"
client = InferenceClient(model=model_name, token=api_token)
messages = [{"role": "system", "content" : "Be nice and provide short but concise answers"},
           {"role": "user", "content": "I'm down bad, what do I do?"}]
```

```python
response = client.chat_completion(messages=messages)
```

```python
reply_text_2 = response.choices[0].content.message
reply_text_2
```
